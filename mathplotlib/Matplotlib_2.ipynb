# Sub-plots 

```python
fig,ax=plt.subplots(figsize=(11,12))

ax.scatter(batters["avg"],batters["strike_rate"],color="green",marker="D")
ax.set_title("Something")
ax.set_xlabel("Avg")
ax.set_ylabel("Strike_rate")
plt.show()
```

```python
plt.scatter(batters["avg"],batters["strike_rate"])
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
ax[0].scatter(batters["avg"],batters["strike_rate"],color="red")
ax[1].scatter(batters["avg"],batters["runs"],color="#E75480")
ax[0].set_title("Batter avg VS strike_rate")
ax[1].set_title("Batter avg VS runs")
ax[0].set_ylabel("strike_rate")
ax[1].set_ylabel("runs")


ax[1].set_xlabel("avg")
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0,0].scatter(batters["avg"],batters["strike_rate"],color="red")
ax[0,1].scatter(batters["avg"],batters["runs"],color="#E75480")
ax[1,0].hist(batters["avg"],color="#E75480")
ax[1,1].hist(batters["runs"],color="#E75480")

```

```python
fig=plt.figure(figsize=(10,6))


ax1=fig.add_subplot(2,2,1)
ax1.scatter(batters["avg"],batters["strike_rate"],color="red")

ax2=fig.add_subplot(2,2,2)
ax2.scatter(batters["avg"],batters["runs"],color="#E75480")

ax3=fig.add_subplot(2,2,3)
ax3.hist(batters["avg"],color="#E75480")

ax4=fig.add_subplot(2,2,4)
ax4.hist(batters["runs"],color="#E75480")
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
xx
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(10,6))
ax=plt.subplot(projection="3d")

p=ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

# Heat map

```python
deliver =pd.read_csv("IPL_Ball_by_Ball_2008_2022.csv")
deliver.head()
```

```python
temp_def=deliver[(deliver["ballnumber"].isin([1,2,3,4,5,6])) & (deliver["batsman_run"]==6)]
grid=temp_def.pivot_table(index="overs",columns="ballnumber",values="batsman_run",aggfunc="count")
grid
```

```python
plt.figure(figsize=(10,20))
plt.imshow(grid)
plt.yticks(deliver["overs"].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
plt.show()
```
